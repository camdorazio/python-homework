```python
#  Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path("budget_data.csv")


# Initialize variables and create a list for each column and assign numerical 
# value to date (i.e. "months") to be able to count
months = 0
pnl = []
dates = []

# Open the csv file as an object
with open(csvpath, "r") as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    # and remove header
    csvreader = csv.reader(csvfile, delimiter=",")
    next(csvreader, None)
    
    # Read each row of data after the header
    for row in csvreader:
       
        # Set the 'Date' and 'Profit/Losses' variables for better
        # readability
        months += 1
        dates.append(row[0])
        pnl.append(int(row[1]))


# Initialize variables
total_pnl = pnl[0]
totaled_delta = 0
greatest_increase = 0
greatest_decrease = 0

# Need to start at 1 rather than 0, since I eliminated the header, so we can calculate the change in profit below 
# I need to put the first month's profit into total profit

for x in range (1, months):
    # Add current month's profit to total profits
    total_pnl += pnl[x]
    
    # Calculate the change/delta between this month and the prior month 
    # and then add it to the totaled change/delta
    current_delta = pnl[x] - pnl[x-1]
    totaled_delta += current_delta
    
    # Calculate the average change/delta. Note: since you are calculating delta's, there is 1 less result in total count
    avg_delta = totaled_delta/(months - 1)
    
    # Use if/else logic to check if the current change/delta was the greatest increase/decrease
    if current_delta > greatest_increase:
        greatest_increase = current_delta
        greatest_increase_date = dates[x]
    elif current_delta < greatest_decrease:
        greatest_decrease = current_delta
        greatest_decrease_date = dates[x]
            
print("Financial Analysis")
print("--------------------------")
print(f"Total Months: {months}")
print(f"Total: ${total_pnl}")
print(f"Average Change: ${avg_delta:.2f}")
print(f"Greatest Increase in Profits: {greatest_increase_date} $({greatest_increase})")
print(f"Greatest Decrease in Profits: {greatest_decrease_date} $({greatest_decrease})")


# Set the path for the pybank_output.txt
output_path = Path("pybank_output.txt")

# Open the output path as a file and write the print statments to the text file. Then close file.

text_file = open("pybank_output.txt", "w")

text_file.write("Financial Analysis\n")
text_file.write("--------------------------\n")
text_file.write(f"Total Months: {months}\n")
text_file.write(f"Total: ${total_pnl}\n")
text_file.write(f"Average Change: ${avg_delta:.2f}\n")
text_file.write(f"Greatest Increase in Profits: {greatest_increase_date} $({greatest_increase})\n")
text_file.write(f"Greatest Decrease in Profits: {greatest_decrease_date} $({greatest_decrease})\n")

text_file.close()

# Set the output header
header = ["Total_Months", "Total_Profit_and_Losses", "Avgerage_Change", "Greatest_Increase_Date", "Greatest_Increase", "Greatest_Decrease_Date", "Greatest_Decrease"]
# Create a list of metrics
metrics = [months, total_pnl, avg_delta, greatest_increase_date, greatest_increase, greatest_decrease_date, greatest_decrease]

# Set the path for the pybank_output.csv
output_path = Path('pybank_output.csv')

# Open the output path as a file and pass into the 'csv.writer()' function
# Set the delimiter/separater as a ','

with open(output_path, 'w') as csvfile:
    # Set the file object as a csvwriter object
    csvwriter = csv.writer(csvfile, delimiter=',')
    # Write the header to the output file
    csvwriter.writerow(header)
    # Write the list of metrics to the output file
    csvwriter.writerow(metrics)
```
